```python
# import xlm package
import xml.etree.ElementTree as ET
import pandas as pd
```

```python
# Parse XML file with annotated laptop reviews


#tree = ET.parse('Laptops_Train.xml')
tree = ET.parse('/Users/gielderks/PycharmProjects/NLP_V2/Other_docs/ABSA16_Laptops_Train_SB1_v2 2.xml')
root = tree.getroot()
```

```python
# Use this dataframe for multilabel classification
# Must use scikitlearn's multilabel binarizer

# Build structured dataframe from xml filie

data = []

labeled_reviews = []
for review in root.findall("Review"):
    
    for sentences in review.findall('sentences'):
        
        for sentence in sentences.findall('sentence'):
            
            entry = []

            if sentence.find('Opinions'):
                
                entry.append(sentence[0].text)
                
                opinions_l = []
                            
                for opinion in sentence.find('Opinions').findall("Opinion"):
                    
                    opinions_l.append(opinion.get("category"))
                
                entry.append(opinions_l)
            
            if entry == []:
                continue
            else:
                data.append(entry)
```

```python
df = pd.DataFrame(data, columns=['ReviewText', 'aspects'])
df['terms'] = None

df.to_csv('Laptop_Train_Data.csv')
df.head(10)
```
